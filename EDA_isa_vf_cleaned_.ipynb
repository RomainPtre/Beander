# Matrice de corr√©lation

```python
# S√©lection des colonnes de note
columns_of_interest = ['Aroma', 'Flavor', 'Aftertaste', 'Acidity', 'Body', 'Balance', 'Uniformity', 'Clean.Cup', 'Sweetness']

# Calcul des corr√©lations
corr_matrix = df[columns_of_interest].corr()

# Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Corr√©lations entre les colonnes sauf les Notes')
plt.show()
```

# Bar plot Acidit√© Maximale par Propri√©taire

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculer l'acidit√© maximale pour chaque propri√©taire
max_acidity_by_owner = df.groupby('Owner')['Acidity'].max().reset_index()

# Cr√©er un bar plot pour visualiser l'acidit√© maximale par propri√©taire
plt.figure(figsize=(80, 20))  # Augmenter la largeur de la figure
sns.barplot(x='Owner', y='Acidity', data=max_acidity_by_owner)
plt.title('Acidit√© Maximale par Propri√©taire')
plt.xlabel('Propri√©taire')
plt.ylabel('Acidit√© Maximale')

# Ajuster les √©tiquettes de l'axe x
plt.xticks(rotation=45, ha='right', fontsize=9)  # Rotation et taille de la police ajust√©es
plt.subplots_adjust(bottom=0.50)  # Ajouter de l'espace en bas pour les √©tiquettes
plt.show()
```

# Bar plot Acidit√© Maximale par Ferme

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculer l'acidit√© maximale pour chaque propri√©taire
max_acidity_by_farm = df.groupby('Farm.Name')['Acidity'].max().reset_index()

# Cr√©er un bar plot pour visualiser l'acidit√© maximale par propri√©taire
plt.figure(figsize=(80, 20))  # Augmenter la largeur de la figure
sns.barplot(x='Farm.Name', y='Acidity', data=max_acidity_by_farm)
plt.title('Acidit√© Maximale par Ferme')
plt.xlabel('Fermes')
plt.ylabel('Acidit√© Maximale')

# Ajuster les √©tiquettes de l'axe x
plt.xticks(rotation=45, ha='right', fontsize=9)  # Rotation et taille de la police ajust√©es
plt.subplots_adjust(bottom=0.50)  # Ajouter de l'espace en bas pour les √©tiquettes
plt.show()
```

# Moyenne total des notes

```python
# Bar plot pour visualiser la moyenne des 'Total.Cup.Points' par 'Variety'
# on observe que l'on a une majorit√© de caf√© de bonne quali√©
plt.figure(figsize=(10, 6))
df.groupby('Variety')['Total.Cup.Points'].mean().plot(kind='bar')
plt.title('Moyenne des Total Cup Points par Vari√©t√©')
plt.xlabel('Vari√©t√©')
plt.ylabel('Moyenne des Total Cup Points')
plt.xticks(rotation=45, ha='right', pad=10)
plt.show()
```
