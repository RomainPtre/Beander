# Matrice de corrélation

```python
# Sélection des colonnes de note
columns_of_interest = ['Aroma', 'Flavor', 'Aftertaste', 'Acidity', 'Body', 'Balance', 'Uniformity', 'Clean.Cup', 'Sweetness']

# Calcul des corrélations
corr_matrix = df[columns_of_interest].corr()

# Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Corrélations entre les colonnes sauf les Notes')
plt.show()
```

# Bar plot Acidité Maximale par Propriétaire

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculer l'acidité maximale pour chaque propriétaire
max_acidity_by_owner = df.groupby('Owner')['Acidity'].max().reset_index()

# Créer un bar plot pour visualiser l'acidité maximale par propriétaire
plt.figure(figsize=(80, 20))  # Augmenter la largeur de la figure
sns.barplot(x='Owner', y='Acidity', data=max_acidity_by_owner)
plt.title('Acidité Maximale par Propriétaire')
plt.xlabel('Propriétaire')
plt.ylabel('Acidité Maximale')

# Ajuster les étiquettes de l'axe x
plt.xticks(rotation=45, ha='right', fontsize=9)  # Rotation et taille de la police ajustées
plt.subplots_adjust(bottom=0.50)  # Ajouter de l'espace en bas pour les étiquettes
plt.show()
```

# Bar plot Acidité Maximale par Ferme

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculer l'acidité maximale pour chaque propriétaire
max_acidity_by_farm = df.groupby('Farm.Name')['Acidity'].max().reset_index()

# Créer un bar plot pour visualiser l'acidité maximale par propriétaire
plt.figure(figsize=(80, 20))  # Augmenter la largeur de la figure
sns.barplot(x='Farm.Name', y='Acidity', data=max_acidity_by_farm)
plt.title('Acidité Maximale par Ferme')
plt.xlabel('Fermes')
plt.ylabel('Acidité Maximale')

# Ajuster les étiquettes de l'axe x
plt.xticks(rotation=45, ha='right', fontsize=9)  # Rotation et taille de la police ajustées
plt.subplots_adjust(bottom=0.50)  # Ajouter de l'espace en bas pour les étiquettes
plt.show()
```

# Moyenne total des notes

```python
# Bar plot pour visualiser la moyenne des 'Total.Cup.Points' par 'Variety'
# on observe que l'on a une majorité de café de bonne qualié
plt.figure(figsize=(10, 6))
df.groupby('Variety')['Total.Cup.Points'].mean().plot(kind='bar')
plt.title('Moyenne des Total Cup Points par Variété')
plt.xlabel('Variété')
plt.ylabel('Moyenne des Total Cup Points')
plt.xticks(rotation=45, ha='right', pad=10)
plt.show()
```
